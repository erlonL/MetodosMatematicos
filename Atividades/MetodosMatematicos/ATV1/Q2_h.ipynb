```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def a(n):
    return n/(np.exp(n))
```

```python
## cabe√ßalho
print(f"{'X':<5}{'Y':<5}")

## elementos
for i in range(nmin, nmax):
    print(f"{i:<5}{a(i):<5}")
```

```python
# dicionario que mapeia os valores de x com y(x)
data = {'x': list(range(nmin, nmax)), 'y': [a(i) for i in range(nmin, nmax)]}

plt.plot(data['x'], data['y'], marker = "o", linestyle = "None", markersize=1)
# plt.yticks([-0.000000001, 0, 0.1])
plt.show()
```

### iv)

```python
nmin = 30
nmax = 40
```

#### Tabela

```python
## cabe√ßalho
print(f"{'X':<5}{'Y':<5}")

## elementos
for i in range(nmin, nmax):
    print(f"{i:<5}{a(i):<5}")
```

#### Figura

```python
L = 0
Epsilon = 10**-3

# dicionario que mapeia os valores de x com y(x)
data = {'x': list(range(nmin, nmax)), 'y': [a(i) for i in range(nmin, nmax)]}

plt.plot(data['x'], data['y'], marker = "o", linestyle = "None")


# LIMITE
plt.plot(data['x'], [L for _ in range(nmin, nmax)])
plt.text(np.mean(data['x']), L + 0.00003, '$L$', ha='center')

# LIMITE INFERIOR
plt.plot(data['x'], [L - Epsilon for _ in range(nmin, nmax)])
plt.text(np.mean(data['x']), L - Epsilon + 0.00003, '$L - \epsilon$', ha='center')

# LIMITE SUPERIOR
plt.plot(data['x'], [L + Epsilon for _ in range(nmin, nmax)])
plt.text(np.mean(data['x']), L + Epsilon + 0.00003, '$L + \epsilon$', ha='center')

plt.yticks([L - Epsilon, L, L + Epsilon])
plt.show()
```
