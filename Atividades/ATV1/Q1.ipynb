```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def a(n):
    return n/(n+1)

# def A(N):
#     return [n / (n + 1) for n in N]
```

```python
nmin = 10
nmax = 100

# print(N, nmin, nmax)
```

## a) o usuário não sabe se a sequência é convergente para um limite L;

### i) Tabela

```python
## cabeçalho
print(f"{'X':<5}{'Y':<5}")

## elementos
for i in range(nmin, nmax):
    print(f"{i:<5}{a(i):<5}")
```

### ii) figura

```python
# dicionario que mapeia os valores de x com y(x)
data = {'x': list(range(nmin, nmax)), 'y': [a(i) for i in range(nmin, nmax)]}

plt.plot(data['x'], data['y'], marker = "o", linestyle = "None", markersize=0.5)
plt.yticks([0, 1, 2])

plt.show()
```

## b) O usuário sabe que a sequência é convergente para um limite L.

```python
L = 1
Epsilon = 10**-1

# dicionario que mapeia os valores de x com y(x)
data = {'x': list(range(nmin, nmax)), 'y': [a(i) for i in range(nmin, nmax)]}

plt.plot(data['x'], data['y'], marker = "o", linestyle = "None", markersize=0.5)


# LIMITE
plt.plot(data['x'], [L for _ in range(nmin, nmax)])
plt.text(np.mean(data['x']), L + 0.002, '$L$', ha='center')

# LIMITE INFERIOR
plt.plot(data['x'], [L - Epsilon for _ in range(nmin, nmax)])
plt.text(np.mean(data['x']), L - Epsilon + 0.002, '$L - \epsilon$', ha='center')

# LIMITE SUPERIOR
plt.plot(data['x'], [L + Epsilon for _ in range(nmin, nmax)])
plt.text(np.mean(data['x']), L + Epsilon + 0.002, '$L + \epsilon$', ha='center')

plt.yticks([L - Epsilon, 0.95, 1, 1.05, L + Epsilon])
plt.show()
```
